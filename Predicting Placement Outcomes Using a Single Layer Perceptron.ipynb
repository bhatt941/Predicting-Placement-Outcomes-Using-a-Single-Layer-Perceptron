```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Asus\Downloads\placement-dataset.csv")
df.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa', y='iq', data = df, hue='placement')
plt.show()
```

```python
x=df[['cgpa']]
y=df['placement']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import Perceptron
pr=Perceptron(alpha=0.01)
```

```python
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100 , pr.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions (x.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```
